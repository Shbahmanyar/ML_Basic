```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.random.rand(300,2)
plt.figure(figsize=(7,3))

plt.scatter (X[:,0],X[:,1],s=20, cmap='summer')
plt.show()
```

```python
k_means = KMeans(n_clusters=3,max_iter=100)
```

```python
k_means.fit(X)
```

```python
plt.figure(figsize=(7,3))

plt.scatter (X[:,0],X[:,1],c= k_means.labels_,s=20, cmap='summer')
plt.scatter (k_means.cluster_centers_[:,0],k_means.cluster_centers_[:,1], marker='x',c='r', s=50, alpha=0.9)
plt.show()
```
